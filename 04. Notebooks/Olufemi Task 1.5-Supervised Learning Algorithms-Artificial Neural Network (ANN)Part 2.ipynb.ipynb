## 1 Import  Libraries & Data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
import operator
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn import metrics
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import multilabel_confusion_matrix, accuracy_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.preprocessing import StandardScaler
```

```python
unscaled_df =pd.read_csv(os.path.join(path, 'Dataset-weather-prediction-dataset-processed.csv'))
```

```python
pleasantdf=pd.read_csv(os.path.join(path, 'Dataset-Answers-Weather_Prediction_Pleasant_Weather.csv'))
```

## 2 Prepare Data

### Remove columns pertaining to weather stations: GDANSK, ROMA, and TOURS

```python
columns = unscaled_df.columns.tolist()
```

```python
len(columns)
```

```python
to_remove = ['GDANSK_cloud_cover',
 'GDANSK_humidity',
 'GDANSK_precipitation',
 'GDANSK_snow_depth',
 'GDANSK_temp_mean',
 'GDANSK_temp_min',
 'GDANSK_temp_max',
'ROMA_cloud_cover',
 'ROMA_wind_speed',
 'ROMA_humidity',
 'ROMA_pressure',
 'ROMA_sunshine',
 'ROMA_temp_mean',
              'TOURS_wind_speed',
 'TOURS_humidity',
 'TOURS_pressure',
 'TOURS_global_radiation',
 'TOURS_precipitation',
 'TOURS_temp_mean',
 'TOURS_temp_min',
            'TOURS_temp_max']
```

```python
for col in to_remove:
    columns.remove(col)
```

```python
len(columns)
```

```python
df = unscaled_df[columns]
```

```python
df.shape
```

```python
df.head()
```

## 3. Split and scale the data

#### Split data

```python
X = df.drop(columns=['DATE', 'MONTH'])
y = pleasantdf.drop(columns=['DATE'])
```

```python
print(X.shape)
print(y.shape)
```

```python
#Split data into training and testing sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Scaling the data
scaler = StandardScaler()

#Fit the train data
scaler.fit(X_train)

#Scaling the X data
X_train_sc = scaler.transform(X_train)
X_test_sc = scaler.transform(X_test)

print(X_train_sc[:10])
```

```python
#Assign the scaled data to DataFrames
X_train_df = pd.DataFrame(X_train_sc, index=X_train.index, columns=X_train.columns)
X_test_df = pd.DataFrame(X_test_sc, index=X_test.index, columns=X_test.columns)
```

```python
X_train_df.head()
```

## 4. ANN on scaled data

### Scenario 1: hidden layer (5, 5), max_iter=500, tol=0.0001

```python
#Create the ANN
#hidden_layer_sizes has up to three layers, each with a number of nodes. So (5, 5) is two hidden layers with 5 nodes each,
#and (100, 50, 25) is three hidden layers with 100, 50, and 25 nodes.
mlp = MLPClassifier(hidden_layer_sizes=(5, 5), max_iter=500, tol=0.0001)
#Fit the data to the model
mlp.fit(X_train_df, y_train)
```

```python
y_pred = mlp.predict(X_train_df)
print(accuracy_score(y_pred, y_train))
y_pred_test = mlp.predict(X_test_df)
print(accuracy_score(y_pred_test, y_test))
```
